```python
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv
load_dotenv()
```

```python
embedding=OpenAIEmbeddings(model='text-embedding-3-large',dimensions=32)
result=embedding.embed_query("Delhi is capital of India")
print(str(result))
```

## For Multiple Documents

```python
documennts=["Delhi is capital of India","Moscow is captial of Russia","Paris is capital of France"]
result= embedding.embed_documents(documennts)

print(str(result))
```

# HuggingFace Embedding Model

```python
from langchain_huggingface import HuggingFaceEmbeddings
embedding=HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
text='Delhi is capital of India'
```

```python
vector=embedding.embed_query(text)
print(str(vector))
```

```python
vector=embedding.embed_documents(documennts)
print(str(vector))
```

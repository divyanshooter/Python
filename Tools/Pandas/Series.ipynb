### Pandas Series

A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

```python
import numpy as np
import pandas as pd
```

## Series from Lists

```python
#creating series from list
country=['India','China','USA','Russia','UK']
pd.Series(country)
```

* Series has value and index
* here values data type is object which is kind of string

runs=[30,99,23,21,45]
pd.Series(runs)

```python
#creating series with custom index
marks=[67,57,89,100]
subjects=['English','Sanskrit','Maths','Hindi']
marks=pd.Series(marks,index=subjects)
```

```python
#setting name
marks=pd.Series(marks,index=subjects,name='12thMarks')
```

## Series from Dictionary

```python
mark={'English':67,'Sanskrit':57,'Maths':89,'Hindi':100}
marks=pd.Series(mark)
```

### Editing Series

```python
#using indexing
marks[1]=98
marks
```

```python
#what if index does not exist and we try to update it, it will create new index
marks['sst']=92
marks
```

```python
#it write time it will write but at read time it will give error
marks['evs']
```

```python
# editing using silicng
runs=pd.Series([10,20,50,10,90])
runs[2:4]=[100,100]
runs
```

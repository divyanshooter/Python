

## 🔁 1. **Anagrams**

* **Definition:** An anagram is a rearrangement of the letters of a word to form a new word (or words).
* **Example:** `"listen"` → `"silent"`

### Python Example:

```python
from itertools import permutations

word = "listen"
anagrams = set([''.join(p) for p in permutations(word)])
print(anagrams)
```

> You can filter these against a dictionary to find real-word anagrams.

---

## 🔠 2. **Bigrams**

* **Definition:** Bigrams are sequences of 2 adjacent items (words or characters).
* **Example (words):** `"I love AI"` → `[('I', 'love'), ('love', 'AI')]`

### Python Example (word-level):

```python
from nltk import bigrams
from nltk.tokenize import word_tokenize

text = "I love AI"
tokens = word_tokenize(text)
bi = list(bigrams(tokens))
print(bi)
```

---

## 🔤 3. **Trigrams**

* **Definition:** Trigrams are sequences of 3 adjacent items.
* **Example:** `"I love deep learning"` → `[('I', 'love', 'deep'), ('love', 'deep', 'learning')]`

### Python Example (word-level):

```python
from nltk import trigrams
from nltk.tokenize import word_tokenize

text = "I love deep learning"
tokens = word_tokenize(text)
tri = list(trigrams(tokens))
print(tri)
```

---

## ⚙️ Optional: Character-level N-grams

If you want bigrams or trigrams of characters instead of words:

```python
from nltk.util import ngrams

text = "hello"
char_bigrams = list(ngrams(text, 2))
char_trigrams = list(ngrams(text, 3))

print("Character bigrams:", char_bigrams)
print("Character trigrams:", char_trigrams)
```

### 🧠 What is **Bag of Words (BoW)**?

**Bag of Words** is a simple and commonly used technique in Natural Language Processing (NLP) to convert text into numerical feature vectors.

* It **ignores grammar and word order** but keeps track of word **frequency**.
* Each document is represented as a **vector of word counts**.

---

### 📦 Example

Suppose you have two texts:

```python
doc1 = "I love AI"
doc2 = "AI loves me"
```

BoW would treat these as:

| Word  | Doc1 | Doc2 |
| ----- | ---- | ---- |
| I     | 1    | 0    |
| love  | 1    | 0    |
| AI    | 1    | 1    |
| loves | 0    | 1    |
| me    | 0    | 1    |

---

### ✅ Implementation in Python (Using `CountVectorizer`)

```python
from sklearn.feature_extraction.text import CountVectorizer

corpus = [
    "I love AI",
    "AI loves me"
]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)

print(vectorizer.get_feature_names_out())
print(X.toarray())
```

**Output:**

```
['ai' 'love' 'loves' 'me']
[[1 1 0 0]
 [1 0 1 1]]
```

---

### 🧰 Notes:

* By default, `CountVectorizer` lowercases and tokenizes on whitespace/punctuation.
* You can add preprocessing steps like stopword removal or stemming.

---

### 🔁 Bag of Words vs TF-IDF

* **BoW** uses raw frequency.
* **TF-IDF (Term Frequency-Inverse Document Frequency)** down-weights common words and emphasizes rare ones.


````python
# BoW

from sklearn.feature_extraction.text import CountVectorizer
count_vectorizer = CountVectorizer(max_features=5000,ngram_range=(1,3))
bag_of_words = count_vectorizer.fit_transform(movies['review'])
bag_of_words = pd.DataFrame(bag_of_words.toarray(),columns = count_vectorizer.get_feature_names_out())
````

````python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_result = pca.fit_transform(bag_of_words.values)
````

# Selecting rows from dataframe


- **iloc** - searches using index positions
- **loc** - searches using index labels 

```python
student_dic={'name':['A','B','C','D'],'iq':[100,90,120,80],'marks':[80,70,100,50],'package':[10,7,14,2]}
students= pd.DataFrame(student_dic)
students.set_index('name',inplace=True)
```

### single row

```python
# this is a series
movies.iloc[0]
```

```python
type(movies.iloc[0])
```

### mupltiple rows

```python
#0 included and 5 excluded
movies.iloc[0:5]
```

```python
# this will give rows from 0 to 5 with alternate rows
movies.iloc[0:5:2]
```

```python
#this will give 5 onwards
movies.iloc[5:]
```

# Filtering the dataframe

```python
ipl[ipl['MatchNumber']=='Final'][['Season','WinningTeam']]
```

```python
#number of time there is superover in ipl
ipl[ipl['SuperOver']=='Y'].shape[0]
```

```python
#number of times CSK wins in Kolkata
ipl[(ipl['WinningTeam']=='Chennai Super Kings') & (ipl['City']=='Kolkata')].shape[0]
```

```python
#toss winner is match winner percentage
(ipl[ipl['WinningTeam']==ipl['TossWinner']].shape[0]/ipl.shape[0] ) *100
```

```python
#movies with imdb rating is higher than 8 and votes>10000
movies[(movies['imdb_votes']>10000) &(movies['imdb_rating']>8)].shape[0]
```

```python
#action movies having rating higher than 7.5
mask=movies['genres'].str.split('|').apply(lambda x:'Action' in x)
movies[(mask) & (movies['imdb_rating']>7.5)]
```

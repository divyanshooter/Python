```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
```

```python
df=pd.read_csv('./data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['last review']=pd.to_datetime(df['last review'],errors='coerce')
```

```python
df.fillna({'last review':df['last review'].min(),'reviews per month':0},inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(subset={'NAME','host name'},inplace=True)
```

```python
df.drop(columns=['house_rules','license'],inplace=True)
```

```python
df['room type']=df['room type'].astype('category')
```

```python
df['price']=df['price'].str.replace('[\$, ]','',regex=True).astype('float')
df['service fee']=df['service fee'].str.replace('[\$, ]','',regex=True).astype('float')
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
sns.displot(df,x='price',kde=True,bins=50)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.show()
```

```python
px.histogram(df,x='price',nbins=50,title='Price Distribution using Plotly Histogram')
```

```python
sns.catplot(data=df,x='room type',kind='count')
```

```python
room_type_count=df['room type'].value_counts().reset_index()
room_type_count.head()
fig=px.pie(room_type_count,names=room_type_count['index'].unique(),values=room_type_count['room type'])
fig.update_traces(textposition='inside',textinfo='percent+label')
fig.show() 
```

```python
plt.figure(figsize=(20,20))
sns.catplot(data=df,y='neighbourhood group',kind='count',order=df['neighbourhood group'].value_counts().index)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df,x='room type',y='price',hue='room type',palette='Set1')
plt.title('Room Type vs Price')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.show()
```

```python
review_over_time=df.groupby(df['last review'].dt.to_period('M')).size()
review_over_time.plot(kind='line')

plt.title('Review over time')
plt.xlabel('Date')
plt.ylabel('No. of Reviews')
```

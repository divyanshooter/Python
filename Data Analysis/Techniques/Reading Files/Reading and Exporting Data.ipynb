# Converters
* It allows you to preprocess data as it’s being loaded, by mapping column names (or indices) to functions.



```python
def rename_teamName(name):
    if name=='Royal Challengers Bangalore':
        return "RCB"
    else:
        return name
```

```python
ipl=pd.read_csv('IPL Matches 2008-2020.csv',parse_dates=['date'],converters={'team1':rename_teamName})
ipl.head()
```

```python
#na_values we can specify which values should be treated as NaN
df=pd.read_csv('aug_train.csv',na_values=['Male'])
df.head()
```
